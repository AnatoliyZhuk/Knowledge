**Случайная величина** - величина, которая в результате испытания примет одно и только одно числовое значение, зависящее от случайных факторов и заранее непредсказуемое.

Задача прогнозирование временного ряда:

**Выборка** (размера T) из временного ряда является подмножеством реализованного пути процесса временного ряда
$x_{1}, x_{2}, x_{3},..., x_{T}$

+ для фиксированного времени временной ряд - это случайная величина
+ для фиксированной реализации временной ряд является функцией времени
+ временной ряд ≈ Стохастический процесс ≈ Стохастическая последовательность


Модель временных рядов для наблюдаемых данных ${x_t}$ - это спецификация совместных распределений последовательности случайных величин ${X_t}$, упорядоченных во времени.

Задача прогнозирования временных рядов - найти функцию $f_T$:<br/>
$y_{T+d}\approx f_T (y_T, ..., y_1, d) \equiv \hat{y}_{T+d|T}$ <br/>
$d \in ({1, ..., D})$ - отсрочка прогноза. Насколько далеко вперед мы хотим предсказывать значение ряда <br/>
$D$ - Горизонт прогнозирования <br/>
При анализе временных рядов, мы надеемся, что данные из прошлого как-то связаны с данными из будущего.

**Тренд** - плавное долгосрочное изменение уровня ряда.<br/>
**Сезонность** - циклические изменения уровня ряда с постоянным периодом.<br/>
**Цикл** - изменения уровня ряда с переменным периодом (встречаются в рядах, связанных с продажами).<br/>
**Ошибка** - непрогнозируемая случайная компонента ряда.<br/>

Квантифицировать степень сходства между значениями ряда в соседних точках можно с помощью **автокорреляции**
$r_{\tau} = \frac{E((y_t - E_y)(y_{t+\tau} - E_y))}{D_y}$ <br/>
$r_{\tau} \in[-1;1]$, $\tau$ - лаг автокорреляции

По своей сути - это обычная корреляция Пирсона между искодным рядом и его версией сдвинутой на несколько отсчетов.<br/>
Количество отсчетов на которое мы сдвигаем ряд называется лагом автокорреляции.<br/>
**Выборочная автокорреляция**
$r_{\tau} = \frac{\sum_{t=1}^{T-\tau}((y_t - \overline{y})(y_{t+\tau} - \overline{y}))}{\sum_{t=1}^{T}(y_t - \overline{y})^2}$ <br/>

Анализировать значения автокорреляции при разных лагах удобно на **коррелограмме** - это график на котором значения автокорреляции отложены в зависимости от $\tau$

+ **IID - independent and identically distributed** (Независимые одинаково распределённые случайные величины) - это ряд, в котором нет тенденции или сезонной составляющей, и в котором наблюдения являются просто независимыми и одинаково распределенными (iid) случайными величинами с нулевым средним.

+ **Слева**: последовательности случайных величин (iid sequences) - **same means same variance**
+ **По центру**: последовательности неслучайных величин (no iid sequences) с изменяющейся во времени дисперсией
+ **Справа**: последовательности неслучайных величин (no iid sequences) с изменяющимся во времени средним

**Инвариантность** - независимость закономерностей от конкретных ситуаций, в которых они устанавливаются, и от способа описания этих ситуаций.<br/>
**Детерминированность** - определямость.<br/>
**Детерминированная система** - это система, в которой случайность не участвует в развитии будущих состояний системы. Таким образом, детерминированная модель всегда будет производить один и тот же результат из заданного начального условия или начального состояния.<br/>
В **детерминированных моделях** выход модели полностью определяется значениями параметров и начальными условиями начальных условий. <br/>
**Стохастическим моделям** присуща некоторая случайность. Один и тот же набор значений параметров и начальных условий приведет к множеству разных выходных данных,то есть последующее состояние такой системы описывается как величинами, которые могут быть предсказаны, так и случайными.


<h2>Стационарность<h2>

**Автоковариация** - это ковариация переменной с самой собой в какое-то другое время, измеренная с помощью временного запаздывания (или опережения) τ.

**Строгая стационарность:** Временной ряд $X_t$ называется строго стационарным, если совместное распределение векторов $(X_1, ..., X_h)$ и $(X_{t + 1}, ..., X_{t + h})$ равны для каждого $t$ и $h$, то есть свойства ряда не зависят от времени.<br/>
Ряды у которых есть тренд (разные средние уровни ряда), сезонность (окно шириной меньше сезонного периода, взятые в разных местах) - нестационарны.<br/>
Ряды с непериодическими циклами являются стационарными (не можем предсказать, где будут находиться максимумы и минимумы ряда).

**Слабая / ковариационная стационарность:** Временной ряд $X_t$ называется слабо стационарным или ковариационно стационарным, если функция среднего $μ_X(t)$ постоянна по $t$ и ковариационная функция $γ_X (t, t + h)$ постоянна по $t$ для каждого $h$.

+ Для слабо стационарных процессов $γ_X (t, t + h) = γ_X(h)$

Формально, чтобы проверить стационарность используют критерий **Дикки-Фуллера**

**Основной момент:** слабая стационарность означает, что среднее значение, дисперсия и автоковариации не меняются со временем. Строгая стационарность подразумевает, что все распределение не меняется со временем.

<h2>Класс прогнозирующих моделей ARMA<h2>

Если сводить задачу прогнозирования временных рядов к задаче обучения с учителем (например сделать регрессию значений ряда на признаки зависящие от времени),то получется плохо, этих признаков недостаточно.

Рассмотрим идею: Что если делать регрессию ряда на собственные значения в прошлом?
$y_t = \alpha + \phi_1 y_{t-1} + \phi_2 y_{t-2} + ... + \phi_p y_{t-p} + \varepsilon_t$

Такая модель называется **моделью авторегрессии порядка $p$ $(AR(p))$**<br/>
$y_t$ - линейная комбинация $p$ предыдущих значений ряда и шумовой компоненты

**Модель скользящее среднее**<br/>
Пусть у нас есть независимый одинаково распределенный во времени шум $\varepsilon_t$<br/>
Для каждого $t$ посчитаем среднее арифметическое между $\varepsilon_t$ и $\varepsilon_{t-1}$
