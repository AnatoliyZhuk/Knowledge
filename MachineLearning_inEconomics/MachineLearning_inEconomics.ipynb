**Временной ряд** - это бесконечная последовательность случайных величин, упорядоченных во времени. Обычно мы предполагаем, что последовательность бесконечна по длине. <br/>
**Временной ряд** - значения признака, измеренные через постоянные временные интервалы.<br/><br/>
$...X_{t-2}, X_{t-1}, X_{t}, X_{t+1}, X_{t+2}...$<br/><br/>

Задача прогнозирование временного ряда:

**Выборка** (размера T) из временного ряда является подмножеством реализованного пути процесса временного ряда
$x_{1}, x_{2}, x_{3},..., x_{T}$

+ для фиксированного времени временной ряд - это случайная величина
+ для фиксированной реализации временной ряд является функцией времени
+ временной ряд ≈ Стохастический процесс ≈ Стохастическая последовательность


Модель временных рядов для наблюдаемых данных ${x_t}$ - это спецификация совместных распределений последовательности случайных величин ${X_t}$, упорядоченных во времени.

+ **IID - independent and identically distributed** (Независимые одинаково распределённые случайные величины) - это ряд, в котором нет тенденции или сезонной составляющей, и в котором наблюдения являются просто независимыми и одинаково распределенными (iid) случайными величинами с нулевым средним.

+ **Слева**: последовательности случайных величин (iid sequences)
+ **По центру**: последовательности неслучайных величин (no iid sequences) с изменяющейся во времени дисперсией
+ **Справа**: последовательности неслучайных величин (no iid sequences) с изменяющимся во времени средним

**Инвариантность** - независимость закономерностей от конкретных ситуаций, в которых они устанавливаются, и от способа описания этих ситуаций.<br/>
**Детерминированность** - определямость.<br/>
**Детерминированная система** - это система, в которой случайность не участвует в развитии будущих состояний системы. Таким образом, детерминированная модель всегда будет производить один и тот же результат из заданного начального условия или начального состояния.<br/>
В **детерминированных моделях** выход модели полностью определяется значениями параметров и начальными условиями начальных условий. **Стохастическим моделям** присуща некоторая случайность. Один и тот же набор значений параметров и начальных условий приведет к множеству разных выходных данных.

Автоковариация - это ковариация переменной с самой собой в какое-то другое время, измеренная с помощью временного запаздывания (или опережения) τ.

**Строгая стационарность:** Временной ряд $X_t$ называется строго стационарным, если совместное распределение векторов $(X_1, ..., X_h)$ и $(X_{t + 1}, ..., X_{t + h})$ равны для каждого $t$ и $h$

**Слабая / ковариационная стационарность:** Временной ряд $X_t$ называется слабо стационарным или ковариационно стационарным, если функция среднего $μ_X(t)$ постоянна по $t$ и ковариационная функция $γ_X (t, t + h)$ постоянна по $t$ для каждого $h$.

+ Для слабо стационарных процессов $γ_X (t, t + h) = γ_X(h)$

**Основной момент:** слабая стационарность означает, что среднее значение, дисперсия и автоковариации не меняются со временем. Строгая стационарность подразумевает, что все распределение не меняется со временем.
