Идеи по улучшению показателей бизнеса берутся из:<br/>
1. Анализа рынка, анализа поведения пользователей и их потребностей
2. Видения как устроен бизнес и как видете его развитие

Для проверки идей существует множество способов:
1. Здравый смысл: обсуждение с коллегами
2. Тестирование идеи на реальных пользователях (стоит дороже - оценка более релевантна)
 + Опрос пользователей
 + Фокус группы - небольшая группа у которой нужно расспрашивать все что они чувствуют по отношению к продукту (Фокус группы глубже по сравнению с опросами, но выборка небольшая)

Требования к проверкам:
1. Достоверность проверки - условия максимально приближены к реальным
2. Минимальные затраты времени и денег




+ Не всегда можно воссоздать с точностью те самые условия, в которых пользователи будут использовать ваш продукт <br/>

**Тестирование сахара в напитке Coca-Cola**<br/>
Была собрана фокус группа - людям предложили 2 напитка: один с обычным содержанием сахара, другой с повышенным.<br/>

* Вывод: условия для фокус группы должны быть максимально приближены к настоящим
* Проблема максимального приближения условий к реальным решается с помощью А/Б - тестирования

**А/Б - тестирование - способ проверки идей в боевых условиях**<br/>
Вы предлагаете пользователям новую функциональность или новый элемент дизайна, именно в тех условиях, в которых пользователи взаимодействуют с вашим продуктом.

 + Изменения в дизайне
 + Изменения в функциональности

+ С одной стороны хочется тратить меньше времени и проверить все идеи сразу <br/> 
+ C другой стороны изменения разных типов плохо тестировать одновременно (может оказаться, что разные изменения технически несовместимы друг с другом) <br/>

Механизм А/Б-тестирования:
1. Планирование эксперимента как именно будет выглядеть А/Б-тест, как делить на группы, сколько будет длиться А/Б-тест
2. Проверка гипотез (как влияет на бизнес)

<h1>Метрики<h1>

+ Выбрать показатель (метрику), который нужно улучшить <br/>
+ Чтобы показать что идея хорошая - нужно провести эксперимент, где применяется эта идея, в которой выбранная метрика должна значимо улучшиться <br/>
+ Метрики могут быть связанны с состоянием бизнеса, с деньгами, аудиторные метрики<br/>
+ Часто бывает, что метрики трудно померить и они могут не реагировать на небольшие изменения в функциональности или дизайне, в других случаях метрики нужно долго измерять<br/>

**Пример:**<br/>
Внеся изменение в ваш сайт с арендой квартир - увеличить количество людей, которые порекомендуют ваш сервис своим знакомым - на измерение этого показателя может понадобиться год (Люди и их знакомые не так часто переезжают) <br/>


Поэтому зачастую полезно использовать прокси-метрики (промежуточные метрики) - метрики, которые с одной стороны достаточно чувствительны, чтобы измерять их в рамках A/B - тестинга, а с другой стороны они согласуются с тем реальным бизнес-показателем, который вы хотите измерять.

В примере аренды квартир в качестве прокси-метрики может быть использована метрика - среднее количество визитов на сайт в день, среднее количество уникальных пользователей (метрики чувствительны, чтобы измерить их в рамках A/B-теста и согласованы с той метрикой, которую мы хотим использовать в реальности)

Важным этапом при принятии решения следует ли запускать ваше решение в A/B-тестинг является **offline тестирование** <br/>
В рамках **offline тестирования** вы можете проверить как ваше изменение сказывается на поведении пользователя по историческим данным <br/>

**Пример:**
Изменение ранжирования результатов по запросам пользователя при поиске квартир на сайте недвижимости.
+ Мы можем проанализировать запросы пользователей за прошлый период
+ Эмулировать новую выдачу новым алгоритмом
+ Зная, что искали пользователи и имея новые результаты поиска - совместив данные, измерим на какие позиции приходятся клики пользователей теперь (алгоритм ранжирования находит то, что пользователи посчитали релевантным - можем сравнить такие метрики, как "средняя позиция клика", если метрика возрастает, то изменение хорошее - имеет смысл протстировать его в онлайне.

Можно заключить, что у нас есть иерархия метрик:
+ Предварительные метрики - мерием до начала эксперимента
+ Экспериментальные метрики - оценивают изменение как хорошее или плохое
+ Бизнес-метрики - надеемся, что наши изменения повлияют на них

<h1>Дизайн эксперимента<h1>

Для того чтобы результаты полученные на небольшой группе можно было обобщать на пользователей в целом - группа должна быть **репрезентативной** (похожая структура - пропорции)<br/><br/>
**При дизайне экспериментов можно либо стратифицировать либо рандомизировать пользователей**<br/>
При построении эксперементальной группы важно выделять какие-то важные свойства пользователей (пол, возраст) и делать так, чтобы в эксперементальной группе были ровно такие же доли по разным подгруппам - такой подход называется **стратификацией**.<br/><br/>

Подход **рандомизации**<br/>
Если вы набираете в вашу группу пользователей абсолютно случайно, то в среднем она получается такого же состава как и вся генеральная совокупность пользователей<br/>
Плюс рандомизации - экспериментальная группа выравнивается со всей генеральной совокупностью, не только по тем показателям, которые вам показались важными, а по всем возможным показателям в целом<br/>


**Группы могут быть независимыми или связанными**<br/>
Связанный (парный) дизайн - ситуация, когда важно померить как на пользователе скажется несколько измерений (выборки связанные).<br/>
Это выгодно когда показатель, который вы измеряете имеет большую индивидуальную дисперсию<br/>
При таком дизайне важно в каком порядке пользователю предлагают разные варианты, чтобы убрать это влияние, можно использовать дизайн крест-накрест (половине пользователей показываете сначала новый вариант, а потом старый, а другим наоборот)<br/>
Экспериментов можно проводить целое количество, однако все хорошо только в том случае, если каждый пользователь участвует только в одном эксперименте.
Если существует вероятность, что каждый пользователь попадает сразу в несколько экспериментов, важно следить, чтобы эксперименты не противоречили друг другу<br/>

Если существует вероятность, что каждый пользователь попадает сразу в несколько экспериментов, важно следить, чтобы эксперименты не противоречили друг другу<br/>
**Пример:** Подбор текста цвета и цвета фона, на котором он показывается не должен быть одинаковым

<h1>Устойчивость<h1>

Устойчивость - одно из главных требований к A/B-тестам<br/>
+ Отсутствие значимых изменений там, где их нет (две одинаковые версии - не должны давать разные значения метрик)
+ Если значимые изменения есть - мы хотим увидеть их на метриках

На практике не всегда так получается<br/>
Бывают ситуации, когда незначимые изменения мы принимаем за значимые (например, если мы изменили цвет кнопки на сайте, возможно часть пользователей просто из любопытства нажмут на нее)

**Обратный эксперимент** - после классического A/B - тестинга, мы продолжаем проводить эксперимент, одноко выделяем маленькую группу пользователей, которая продолжает видеть старое решение (помогает убедиться, что наши изменения действительно зачимы)

+ На обратный эксперимент уходит время и если по итогу его проведения, мы понимаем, что лучше ситуация не стала, то это нежелательная для нас ситуация - для того, чтобы избежать такой ситуации, существует **технология A/A-тестинга**

Делим на группы как и при A/B - тестинге, только будем показывать разным группам один и тот же алгоритм, не будем в тестовой группе показывать наше новое решение
+ Если мы не видим значимые изменения - хорошая ситуация
+ Если мы видим значимые изменения - что-то не так в дизайне эксперимента

**Как оценить размер выборки и длительность эксперимента?**

<h1>Параметрические критерии проверки гипотез<h1>

Параметрические критерии предполагают, что выборки с которыми они имеют дело взяты из конкретных распределений и проверяют гипотезы, связанные со значениями параметров этих распределений.

<h2>Одновыборочные критерии Стьюдента<h2>

Начнем с критериев, которые связаны с гипотезами о значениях параметров нормального распределения (семейство критериев Стюдента - позволяют проверять гипотезы о математических ожиданиях нормальных распределений).


Пример: **Вес детей при рождении**<br/>
Средний вес детей при рождении - 3.3 кг<br/>
У женщин, живущих за чертой бедности - 2.8 кг <br/>
Мы заинтересованы, чтобы у женщин за чертой бедности дети рождались тяжелее <br/>
Поэтому была разработана эксперементальная программа ведения беременности, чтобы проверить ее эффективность проводится эксперимент, в котором принимает участие 25 женщин, живущих за чертой бедности. <br/>
Средний вес их дитей при рождении - 3075г, стандартное отклонение - 500г <br/>
**Эффективна ли эта программа?**


Введем математический аппарат:<br/>
Выборка из нормального распределения: $X^n = (x_1, ..., X_n)$<br/>
$X \sim N(\mu, \sigma^2)$ $\sigma$ - известна<br/>
Мы хотим проверить гипотезу $H_0: \mu$ (среднее значение нашего распределения) = $\mu_{0}$  (какая-то константа)<br/>
Альтернативная гипотеза $H_1: \mu \ne \mu_{0}$ - это двусторонняя альтернатива <br/>
Если бы мы проверяли альтернативы $H_1: \mu <> \mu_{0}$ - это одностороние альтернативы<br/>
Нулевую гипотезу мы будем проверять с помощью $Z$-статистики:<br/>
$Z(X^n) = \frac{\overline{X} - \mu_{0}}{\sigma / \sqrt{n}}$<br/>
$\overline{X}$ - среднее выборочное значение <br/>
Если нулевая гипотеза справедлива, то наша статистика имеет стандартное нормальное распределение<br/>
$X \sim N(0, 1)$<br/>
Большая часть критериев, которые имеют стандартное нормальное распределение называются $Z$-критериями

Как посчитать достигаемый уровень значимости такого критерия, зависит от альтернатив, которые мы используем.<br/>


Рассмотрим односторонюю альтернативу: $H_1: \mu < \mu_{0}$<br/>
Если такая альтернатива справедлива, то более вероятными являются маленькие значения $Z$-статистики<br/>
Допустим мы получили значение $Z=-2$<br/>
Тогда достигаемый уровень значимости - это интеграл от плотности нормального распределения от $-\infty$ до $-2$<br/>
На самом деле, само значение интеграла вычислять не нужно, потому что значение этого интеграла = значению функции нормального стандартного распределения в точке $Z$<br/>
$p = F_{N(0, 1)}(z)$

Рассмотрим односторонюю альтернативу: $H_1: \mu > \mu_{0}$<br/>
Более вероятные большие значения статистики<br/>
Чтобы посчитать уровень значимости, нужно посчитать интеграл по правому хвосту плотности нашего нулевого распределения<br/>
Достигаемый уровень значимости - это интеграл от плотности нормального распределения от $-2$ до $\infty$<br/>
$p = 1 - F_{N(0, 1)}(z)$

image.png

$H_1: \mu \ne \mu_{0}$<br/>
Если используется двустороняя альтернатива, то при ее справедливости более вероятными будут большие по модулю значения статистики $Z$<br/>
При подсчете достигаемого уровня значимости вас будет интересовать и левый и правый хвосты нашего нулевого распределения<br/>
$p = 2*(1 - F_{N(0, 1)}(|z|))$

image.png

Теперь давайте разберем случай, когда дисперсия неизвестна<br/>
В таком случае, вместо $Z$-критерия Стьюдента нужно использовать $t$-критерий Стюдента<br/>
Заменить $\sigma$ на $S$

Выборка из нормального распределения: $X^n = (x_1, ..., X_n)$<br/>
$X \sim N(\mu, \sigma^2)$ $\sigma$ - известна<br/>
Гипотеза $H_0: \mu$ (среднее значение нашего распределения) = $\mu_{0}$  (какая-то константа)<br/>
Альтернативная гипотеза $H_1: \mu \ne \mu_{0}$ - это двусторонняя альтернатива <br/>

$T(X^n) = \frac{\overline{X} - \mu_{0}}{S / \sqrt{n}}$<br/>
$\overline{X}$ - среднее выборочное значение <br/>
Если нулевая гипотеза справедлива, то наша статистика имеет распределение Стьюдента с числом степеней свободы $St(N-1)$<br/>


Достигаемый уровень значимости для $t$-критерия Стьюдента считается также как и для $Z$-критерия <br/>
$p = F_{St(N-1)}(t)$ при $H_1: \mu > \mu_{0}$<br/>
$p = 1 - F_{St(N-1)}(t)$ при $H_1: \mu < \mu_{0}$ <br/>
$p = 2*(1 - F_{St(N-1)}(|t|))$ при $H_1: \mu \ne \mu_{0}$<br/>

Единственное отличие заключается в том, что вместо функции нормального распределения вы берете функцию распределения Стьюдента с числом степеней свободы $N-1$<br/>
С ростом объема выборки разница между $t$ и $Z$ критериями уменьшается<br/>
Чем больше наблюдений, тем точнее выборочная дисперсия $S^2$ оценивает теоретическую дисперсию $\sigma^2$<br/>
С ростом количества наблюдений увеличивается число степеней свободы нулевого распределения $T$-критерия, а чем больше степеней свободы у распределения Стьюдента, тем больше оно похоже на стандартное нормальное (больше 30 наблюдений распределения почти не отличимы)<br/>

**Возвращаемся к задаче о весе детей**<br/>
$H_0$: программа не эффективна $\mu$  = 2800<br/>
Альтернативная гипотеза $H_1$: программа как-то влияет $\mu \ne 2800$<br/>
$t$-критерий Стьюдента показывает: $p$=0.0111<br/>
Средний вес детей увеличивается на 275г (95% доверительный интервал - [68.6, 481.4])<br/>
$\bar{X} \pm t_{n-1, 1-0.95/2}*\frac{S}{\sqrt{n
}}$<br/>
$275 \pm t_{24, 0.525}*\frac{500}{5} = 275 \pm 2.064 * 100 = 275 \pm 206.4$<br/>
Получаем интервал [68.6, 481.4]<br/>
Отвергаем нулевую гипотезу, так как ноль не попадает в интервал.

Для односторонней альтернативы<br/>
$H_0$: программа не эффективна $\mu$  = 2800<br/>
Альтернативная гипотеза $H_1$: программа как-то влияет $\mu > 2800$<br/>
Посчитаем нижний интервал:<br/>
$275 \pm t_{24, 0.05}*\frac{500}{5} = 275 \pm 1.711 * 100 = 275 \pm 171.1 = 103.9$<br/>
Отвергаем нулевую гипотезу, так как ноль не попадает в интервал.

Используя одностороннюю альтернативу вместо двухсторонней, можно получить достигаемый уровень значимости в два раза меньше, иногда это помогает получить совсем другой результат.<br/>
Использовать одностороннюю альтернативу можно если:
